```python
import pandas as pd


df = pd.read_csv('/content/drive/MyDrive/Titanic-Dataset.csv')

total_fare = df['Fare'].sum()

first_class_fare = df[df['Pclass'] == 1]['Fare'].sum()

second_class_fare = df[df['Pclass'] == 2]['Fare'].sum()
third_class_fare = df[df['Pclass'] == 3]['Fare'].sum()

class_fares = [first_class_fare, second_class_fare, third_class_fare]

class_percentages = [(fare / total_fare) * 100 for fare in class_fares]


print(f"Percentage of total fare revenue by class:")
print(f"First Class: {class_percentages[0]:.2f}%")
print(f"Second Class: {class_percentages[1]:.2f}%")
print(f"Third Class: {class_percentages[2]:.2f}%")
```

```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/Titanic-Dataset.csv')

def categorize_age(age):
    if age < 18:
        return 'child'
    elif 18 <= age < 65:
        return 'adult'
    else:
        return 'senior'

df['age_group'] = df['Age'].apply(categorize_age)

total_passengers = len(df)

age_group_counts = df['age_group'].value_counts()

age_group_proportions = age_group_counts / total_passengers

age_group_percentages = age_group_proportions * 100

print(age_group_percentages)
```

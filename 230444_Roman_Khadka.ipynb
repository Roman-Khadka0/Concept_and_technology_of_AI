Task - 3 - Outlier Detection:
Tasks:
1. Identify outlier countries in South Asia based on their Score and GDP per Capita.
2. Define outliers using the 1.5 × IQR rule.
3. Create a scatter plot with GDP per Capita on the x-axis and Score on the y-axis, highlighting outliers
in a different color.
4. Discuss the characteristics of these outliers and their potential impact on regional averages.

```python
#1
#Identifying outlier countries in South Asia based on their Score and GDP per Capita.

# Calculating IQR for Score and GDP per Capita
Q1_score = south_asian_df['score'].quantile(0.25)
Q3_score = south_asian_df['score'].quantile(0.75)
IQR_score = Q3_score - Q1_score

Q1_gdp = south_asian_df['Log GDP per capita'].quantile(0.25)
Q3_gdp = south_asian_df['Log GDP per capita'].quantile(0.75)
IQR_gdp = Q3_gdp - Q1_gdp
```

```python
#2
# Defining outlier thresholds
lower_bound_score = Q1_score - 1.5 * IQR_score
upper_bound_score = Q3_score + 1.5 * IQR_score

lower_bound_gdp = Q1_gdp - 1.5 * IQR_gdp
upper_bound_gdp = Q3_gdp + 1.5 * IQR_gdp

# Identifying outliers
outliers = south_asian_df[
    (south_asian_df['score'] < lower_bound_score) | (south_asian_df['score'] > upper_bound_score) |
    (south_asian_df['Log GDP per capita'] < lower_bound_gdp) | (south_asian_df['Log GDP per capita'] > upper_bound_gdp)
]
```

```python
#3
# Create the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(south_asian_df['Log GDP per capita'], south_asian_df['score'], label='Countries', alpha=0.7)
plt.scatter(outliers['Log GDP per capita'], outliers['score'], color='red', label='Outliers', s=100)  # Highlight outliers

plt.xlabel('Log GDP per Capita')
plt.ylabel('Happiness Score')
plt.title('Outlier Detection in South Asia')
plt.legend()

# Add country labels to the plot
for i in range(len(south_asian_df)):
    country = south_asian_df.iloc[i]['Country name']
    gdp = south_asian_df.iloc[i]['Log GDP per capita']
    score = south_asian_df.iloc[i]['score']
    plt.annotate(country, (gdp, score), fontsize=8)


plt.show()

# Display the outlier countries
print("\nOutlier Countries:")
display(outliers[['Country name', 'score', 'Log GDP per capita']])
```

as per the information; the outlier nations tend to be those with much lower happiness rankings than the rest of the region.  These nations may be dealing with social problems, political unrest, or economic difficulties (as seen by GDP per capita figures that may be below the lower bound of the IQR).

#Impact on regional averages:
Outliers have a chance to affect regional values, especially the mean. Outliers have a capacity to inflate or deflate South Asia's GDP per capita and overall average happiness score, which could be a false representation of the region's normal situation because the mean is sensitive to extreme values.  In this case, the median would be a more trustworthy measure of central tendency because it is less impacted by outliers.



# Further Investigation:
For a deeper understanding of the characteristics of outlier countries, other factors such as social support, healthy life expectancy, freedom to make life choices, generosity, and views of corruption could be investigated. Analyzing such variables may identify the root cause for the outlier status. This could include additional statistical studies and testing of predictions.  Additional data from other sources may also be useful for validating the conclusions and understanding previous patterns.

Task - 4 - Exploring Trends Across Metrics:
Tasks:
1. Choose two metrics (e.g., Freedom to Make Life Choices and Generosity) and calculate their correlation
{pearson correlation} with the Score for South Asian countries.
2. Create scatter plots with trendlines for these metrics against the Score.
3. Identify and discuss the strongest and weakest relationships between these metrics and the Score for
South Asian countries.

```python
# 1. Calculating Pearson correlation
metric1 = 'Freedom to make life choices'
metric2 = 'Generosity'

# Calculating Pearson correlation
correlation1 = south_asian_df['score'].corr(south_asian_df[metric1])
correlation2 = south_asian_df['score'].corr(south_asian_df[metric2])

# Displaying the results
print(f"Pearson correlation between Score and {metric1}: {correlation1:.2f}")
print(f"Pearson correlation between Score and {metric2}: {correlation2:.2f}")
```

```python
def scatter_with_trendline(x, y, x_label, y_label, title):
    # Create a DataFrame for easier plotting with Seaborn
    data = pd.DataFrame({x_label: x, y_label: y})

    # Create scatter plot with trendline
    plt.figure(figsize=(12, 6))
    sns.regplot(
        x=x_label,
        y=y_label,
        data=data,
        scatter_kws={'alpha': 0.7},  # Transparency of scatter points
        line_kws={'color': 'red'},   # Color of trendline
    )

    # Add labels and title
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(title)
    plt.grid(True)
    plt.show()

# Create scatter plots with trendlines
scatter_with_trendline(
    south_asian_df['score'],
    south_asian_df[metric1],
    'Score',
    metric1,
    f'Relationship between Score and {metric1}\n'
)

scatter_with_trendline(
    south_asian_df['score'],
    south_asian_df[metric2],
    'Score',
    metric2,
    f'Relationship between Score and {metric2}'
)
```

```python
# 3. Discuss relationships
print("\nDiscussion:")
if abs(correlation1) > abs(correlation2):
    print(f"The strongest relationship is between Score and {metric1}, with a correlation of {correlation1:.2f}.")
    print(f"The weakest relationship is between Score and {metric2}, with a correlation of {correlation2:.2f}.")
else:
    print(f"The strongest relationship is between Score and {metric2}, with a correlation of {correlation2:.2f}.")
    print(f"The weakest relationship is between Score and {metric1}, with a correlation of {correlation1:.2f}.")
```

```python
# Calculating IQR for outlier detection
def find_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers
```

```python
# South Asia Outliers
south_asia_outliers_score = find_outliers_iqr(south_asian_df, 'score')
south_asia_outliers_gdp = find_outliers_iqr(south_asian_df, 'Log GDP per capita')

# Combineing and display outlier information
print("South Asia Outliers (Score):\n")
display(south_asia_outliers_score[['Country name', 'score']])
print("\nSouth Asia Outliers (GDP per Capita):\n")
display(south_asia_outliers_gdp[['Country name', 'Log GDP per capita']])
```

```python
# Middle East Outliers
middle_east_outliers_score = find_outliers_iqr(middle_eastern_df, 'score')
middle_east_outliers_gdp = find_outliers_iqr(middle_eastern_df, 'Log GDP per capita')

# Combining and display outlier information
print("\nMiddle East Outliers (Score):\n", middle_east_outliers_score[['Country name', 'score']])
print("\nMiddle East Outliers (GDP per Capita):\n",middle_east_outliers_gdp[['Country name', 'Log GDP per capita']])
```
